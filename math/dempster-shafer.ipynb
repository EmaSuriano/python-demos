## Calculo de Dempster-Shafer

> https://en.wikipedia.org/wiki/Dempster%E2%80%93Shafer_theory


```python
masas = {
    "A": 0.7,
    "B": 0.05,
    "C": 0.02,
    "D": 0.03,
    "AB": 0.1,
    "AC": 0,
    "AD": 0.05,
    "BC": 0,
    "BD": 0,
    "CD": 0,
    "ABC": 0,
    "ABD": 0.05,
    "ACD": 0,
    "BCD": 0,
    "ABCD": 0,
}

# basic check to avoid typos when loading weights
assert sum(masas.values()) == 1
```

```python
import itertools

creencias = {}


def all_combinations(arr: list[str]) -> list[str]:
    res = []

    for i in range(1, len(arr) + 1):
        keys = ["".join(k) for k in itertools.combinations(arr, i)]
        res.extend(keys)

    return res


for i in masas:
    combinations = all_combinations(list(i))
    res = sum(map(lambda key: masas[key], combinations))
    creencias[i] = round(res, 2)
```

```python
plausibilidad = {}

for i in masas:
    keys = [k for k in masas if not any(j in k for j in list(i))]
    res = 1 - sum(map(lambda key: masas[key], keys))
    plausibilidad[i] = round(res, 2)
```

```python
incertidumbre = {}

for i in masas:
    incertidumbre[i] = round(plausibilidad[i] - creencias[i], 2)
```

```python
import pandas as pd

df = pd.DataFrame(
    {"m": masas, "Bel": creencias, "Pl": creencias, "Inc": creencias},
    index=masas.keys(),
)

display(df)
```
