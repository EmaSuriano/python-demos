```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
# Load initial image --> the default
img = cv2.imread("./Lena_RGB.png")
print("Image dimensions: ", np.shape(img))

# Convert BGR en RGB
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_RGB)
plt.show()
```

## RGB channels


```python
fig, ax = plt.subplots(1, 3)
channels = cv2.split(img)

for i, value in enumerate(channels):
    ax[i].imshow(value, cmap="gray")
    ax[i].set_title("RBG"[i])

plt.show()
```

## Brightness


```python
brightness_value = 50  # Positive for brighter, negative for darker


# Display the original, brighter, and darker images
fig, ax = plt.subplots(1, 3)

ax[0].imshow(img_RGB)
ax[0].set_title("Original Image")


brighter_image = cv2.add(img, np.ones(img.shape, dtype=np.uint8) * brightness_value)
brighter_image_rgb = cv2.cvtColor(brighter_image, cv2.COLOR_BGR2RGB)
ax[1].imshow(brighter_image_rgb)
ax[1].set_title("Brighter Image")

darker_image = cv2.subtract(img, np.ones(img.shape, dtype=np.uint8) * brightness_value)
darker_image_rgb = cv2.cvtColor(darker_image, cv2.COLOR_BGR2RGB)
ax[2].imshow(darker_image_rgb)
ax[2].set_title("Darker Image")

plt.show()
```

## Contrast


```python
fig, ax = plt.subplots(1, 3)


ax[0].imshow(img_RGB)
ax[0].set_title("Original Image")

contrasted_image = cv2.convertScaleAbs(img_RGB, alpha=1.5, beta=0)
contrasted_image_rgb = cv2.cvtColor(contrasted_image, cv2.COLOR_BGR2RGB)
ax[1].imshow(contrasted_image)
ax[1].set_title("More Contrast")

contrasted_image = cv2.convertScaleAbs(img_RGB, alpha=0.5, beta=0)
contrasted_image_rgb = cv2.cvtColor(contrasted_image, cv2.COLOR_BGR2RGB)
ax[2].imshow(contrasted_image)
ax[2].set_title("Less Contrast")

plt.show()
```

## Cropping


```python
figs, ax = plt.subplots(1, 2)

ax[0].imshow(img_RGB)
ax[0].set_title("Original")

new_img = img_RGB[90:190, 20:305]
ax[1].imshow(new_img)
ax[1].set_title("Crop")

plt.show()
```

## Rotation


```python
figs, ax = plt.subplots(1, 2)
ax[0].imshow(img_RGB)
ax[0].set_title("Original")

rows, cols, ch = img_RGB.shape

M = cv2.getRotationMatrix2D(
    (cols / 2, rows / 2),
    angle=30,
    scale=1.4,
)
new_img = cv2.warpAffine(img_RGB, M, (cols, rows))  # Aplico la transformación
ax[1].imshow(new_img, cmap="gray")
ax[1].set_title("rotación")

plt.show()
```

## Flipping


```python
figs, ax = plt.subplots(1, 4)

ax[0].imshow(img_RGB, cmap="gray")
ax[0].set_title("original")

flipVertical = cv2.flip(img_RGB, 0)
ax[1].imshow(flipVertical, cmap="gray")
ax[1].set_title("flip vertical")

flipHorizontal = cv2.flip(img_RGB, 1)
ax[2].imshow(flipHorizontal, cmap="gray")
ax[2].set_title("flip horizontal")

flipBoth = cv2.flip(img_RGB, -1)
ax[3].imshow(flipBoth, cmap="gray")
ax[3].set_title("flip ambos")

plt.show()
```
